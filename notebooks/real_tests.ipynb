```python
import matplotlib.pyplot as plt
import sys
import os

# Get the project root (one folder up from notebooks/)
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
src_path = os.path.join(project_root, 'src')

# Add to Python path
sys.path.append(src_path)



from crash_signature import (
    load_multiple_summaries,
    build_crash_signature,
    classify_crash,
    infer_tail_key,
)

crash_ids = ["Crash1", "Crash2", "Crash3"]
summaries = load_multiple_summaries("../results", crash_ids)
sum1, sum2, sum3 = summaries

print("Crash1 summary:", sum1)
print("Crash2 summary:", sum2)
print("Crash3 summary:", sum3)

# Build signature from Crash1 & Crash2
signature = build_crash_signature(sum1, sum2)
print("Signature:", signature)

# Classify all three against that signature
cls1 = classify_crash(sum1, signature)
cls2 = classify_crash(sum2, signature)
cls3 = classify_crash(sum3, signature)

print("Crash1 vs signature:", cls1)
print("Crash2 vs signature:", cls2)
print("Crash3 vs signature:", cls3)
```

```python
# Build signature using new feature-based method
tail_key = infer_tail_key(sum1)
feat_sig = build_feature_signature(sum1, sum2, tail_key=tail_key, use_std=True)

# Compute distances of Crash1,Crash2,Crash3 to center
from crash_signature import compute_feature_vector, feature_distance

f1 = compute_feature_vector(sum1, tail_key)
f2 = compute_feature_vector(sum2, tail_key)
f3 = compute_feature_vector(sum3, tail_key)
center = feat_sig["center"]

d1 = feature_distance(f1, center)
d2 = feature_distance(f2, center)
d3 = feature_distance(f3, center)

print("Distances:", d1, d2, d3)

# A simple threshold: max(d1, d2) * 1.5
thr = max(d1, d2) * 1.5
print("Proposed threshold:", thr)
print("Crash3 within crash-regime?", d3 <= thr)
```
